```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Дані
x1 = np.array([0, 0, 0, 1, 1, 2, 2, 2])
x2 = np.array([1.5, 2.5, 3.5, 1.5, 3.5, 1.5, 2.5, 2.5])
y = np.array([2.3, 8.5, 0.5, 2, 1, 9.1, 5, 7.2])

# Формуємо матрицю X (додаємо стовпець одиниць для вільного члена a0)
X = np.column_stack((np.ones(len(x1)), x1, x2))

# Метод найменших квадратів
model = sm.OLS(y, X).fit()
a0, a1, a2 = model.params

# Виведення рівняння регресії
print(f"Рівняння: y = {a0:.4f} + {a1:.4f} * x1 + {a2:.4f} * x2")
```

```python

# Обчислення значення функції у точці (x1=1.5, x2=3)
x1_new, x2_new = 1.5, 3
y_pred = a0 + a1 * x1_new + a2 * x2_new
print(f"Значення функції у точці (1.5, 3): y = {y_pred:.4f}")
```

```python

# Коефіцієнт детермінації R^2
r2 = model.rsquared
print(f"Коефіцієнт детермінації R^2 = {r2:.4f}")
```

```python

# Побудова графіка
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y, color='red', label='Дані')

# Побудова площини регресії
x1_range = np.linspace(min(x1), max(x1), 10)
x2_range = np.linspace(min(x2), max(x2), 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
y_grid = a0 + a1 * x1_grid + a2 * x2_grid
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.8, color='cyan')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('Модель лінійної регресії')
plt.legend()
plt.show()
```
